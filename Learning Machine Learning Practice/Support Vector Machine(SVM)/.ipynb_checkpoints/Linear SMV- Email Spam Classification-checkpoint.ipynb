```python
import numpy as pn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import validation_curve
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
email = pd.read_csv("Data/Spam.csv", sep=',', header = None)
email.head()
```

```python
email.columns = email.iloc[0][:]
```

```python
email.head()
```

```python
email.drop(email.index[0],axis = 0, inplace=True)
```

```python
email.head()
```

```python
email.info()
```

```python
email.isnull().sum()
```

```python
email['spam'].describe()
```

```python
x = email.drop('spam', axis=1)
y = email.spam.values.astype(int)
```

```python
from sklearn.preprocessing import scale
x = scale(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=4)
```

```python
print(y_train.mean())
print(y_train.mean())
```

```python
model = SVC(C = 1)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
confusion_matrix(y_true=y_test, y_pred=y_pred)
```

```python
from sklearn import metrics
print('Accuracy: ', metrics.accuracy_score(y_test, y_pred))
print('Precision: ', metrics.precision_score(y_test, y_pred))
print('Recall: ', metrics.recall_score(y_test, y_pred))
```

```python
folds = KFold(n_splits = 5, shuffle=True, random_state = 4)
model = SVC(C=1)
```

```python
cv_results = cross_val_score(model, x_train, y_train, cv=folds, scoring='accuracy')
```

```python
print(cv_results)
print('Mean Accuracy: {}'.format(cv_results.mean()))
```

```python
params = {'C': [0.1,1,10,100,1000]}
model = SVC()
model_cv = GridSearchCV(estimator = model, param_grid = params,scoring='accuracy', cv = folds,verbose = 1, return_train_score=True)
```

```python
model_cv.fit(x_train,y_train)
```

```python
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
plt.figure(figsize=(8,6))
plt.plot(cv_results['param_C'],cv_results['mean_test_score'])
plt.plot(cv_results['param_C'], cv_results['mean_train_score'])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.legend(['Test Accuracy', 'Train_Accuracy'], loc='upper left')
plt.xscale('log')
```

```python
best_score = model_cv.best_score_
best_C = model_cv.best_params_['C']
print('The height test accuracy is {0} at c = {1}'.format(best_score, best_C))
```

```python
model = SVC(C=best_C)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
print('Accuracy: ', metrics.accuracy_score(y_test, y_pred))
print('Precision: ', metrics.precision_score(y_test, y_pred))
print('Recall: ', metrics.recall_score(y_test, y_pred))
```

```python
params = {"C": [0.1, 1, 10, 100, 1000]}

# specify scores/metrics in an iterable
scores = ['accuracy', 'precision', 'recall']

for score in scores:
    print("# Tuning hyper-parameters for {}".format(score))
    
    clf = GridSearchCV(SVC(), 
                       params, 
                       cv=folds,
                       scoring=score,
                       return_train_score=True)
    clf.fit(x_train, y_train)

    print(" The highest {0} score is {1} at C = {2}".format(score, clf.best_score_, clf.best_params_))
    print("\n")
```
