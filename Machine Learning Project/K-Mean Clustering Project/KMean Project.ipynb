```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Generate Data - two means
mean1 = [np.random.randint(50),np.random.randint(50)]
mean2 = [np.random.randint(50),np.random.randint(50)]

#make a diagonal covarianc
cov = [[100,0],[0,100]]

x1,y1 = np.random.multivariate_normal(mean1,cov,100).T
x2,y2 = np.random.multivariate_normal(mean2,cov,100).T

x = np.append(x1,x2)
y = np.append(y1,y2)


plt.plot(x,y,"x")
plt.show()
```

```python
X = np.array(list(zip(x,y)))
```

```python
#Make a KMean model
kmeans = KMeans(n_clusters = 2)
```

```python
#fit the model of data
kmeans.fit(X)
```

```python
centroids = kmeans.cluster_centers_
label = kmeans.labels_

print(centroids)
print(label)
```

```python
colors = ['g.','r.']

for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colors[label[i]], markersize=10)
plt.scatter(centroids[:,0],centroids[:,1], marker = 'x',zorder=10)
plt.show()
```
